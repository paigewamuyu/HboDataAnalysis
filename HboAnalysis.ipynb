```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Import the datasets
df1= pd.read_csv('hbo_max_content.csv')
df1.head()
```

```python
df2= pd.read_csv('hbo_content.csv')
df2.head()
```

```python
#merge the two datasets to generate a new dataset named df
df = df1.merge(df2, how= 'inner', on='index')
df
```

```python
#new dataset summary
df.info()
```

```python
#find missing values in each column
missing_values = df.isnull().sum()
missing_values
```

```python
# drop the missing values(containing NaN)
df_dropped = df.dropna()
df_dropped
```

```python
#find any duplicates in the dataset after dropping rows containing NaN values
duplicates = df_dropped.duplicated().sum()
duplicates
```

```python
#confirm whether all rows containing missing values have been dropped
df_null = df_dropped.isna().sum()
df_null
```

```python
df_no_duplicates = df_dropped.drop_duplicates()
df_no_duplicates 
```

```python
#converts the newly generated dataset into a csv file named clean data
df_no_duplicates.to_csv('clean_data.csv')
```

```python
#load the new csv file
df = pd.read_csv ('clean_data.csv')
df
```

```python
# summary statistics for the DataFrame
df.describe(include='all').T
```

```python
print(df.columns)
```

```python
#Data visualisations

# barplot for the IMDB score and the corresponding movie title
fig, ax = plt.subplots(figsize = (12,5))
sns.barplot(data= df, x='title_x', y='imdb_score_x',width=0.4)
plt.title('IMDB Score and Title',size= 25)
plt.xlabel('Title',size= 25)
plt.ylabel('IMDB Score',size= 25)
 # Rotate x-axis labels for better visibility
plt.xticks(rotation=90)
plt.show()


# barplot for the IMDB score and the corresponding movie title
fig, ax = plt.subplots(figsize = (12,5))
sns.barplot(data= df, x='title_x', y='rotten_score_x',width=0.4)
plt.title('Rotten Score and Title',size= 25)
plt.xlabel('Title',size= 20)
plt.ylabel('Rotten Score',size= 20)
plt.xticks(rotation=90)
plt.show()
```

The interpretation of this bar plot would be that the IMDB score of a TV show or movie varies with the year it was released. The plot shows that the IMDB score is higher for a particular year, it means that the average IMDB score of TV shows or movies released in that year was higher than the average IMDB score of TV shows or movies released in other years.

```python
#separating numericl and categorical variables for analysis
cat_cols=df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
#generate a series of histograms and boxplots for each numerical variable in the dataframe.
for col in num_cols:
    print(col)
    print('Skew :', round(df[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    df[col].hist(grid=True)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.show()
```

So, for each numerical column in the dataframe, this code prints the column name, calculates and prints its skewness, and displays a histogram and a boxplot of the column's values.

```python
# histogram for the movie/series frequency and its IMDB score
df['imdb_score_x'].hist(bins=20, edgecolor='black')
plt.title('Histogram of imdb score')
plt.xlabel('imdb score')
plt.ylabel('Frequency')
plt.show()
```

```python
#boxplot representing the year and corresponding IMDB score
plt.figure(figsize=(10, 6))
sns.boxplot(x='year_x', y='imdb_score_x', data=df)
plt.title('Boxplot of IMDB Score by Year')
plt.show()
```

The line inside the box shows the median of IMDB score for each year. The box shows the interquartile range (IQR), which is the range between the 25th and 75th percentiles. This gives you a sense of the middle 50% of scores. The whiskers which extend from the box to show the range of the data. By default, they extend no more than 1.5 * IQR from the edges of the box, ending at the furthest data point within that interval. The outliers are plotted as individual points. These are data points that are more than 1.5 * IQR from the box. By comparing the boxplots for different years, you can see how the distribution of IMDB score has changed over time.

```python
#Separate the data into subsets based on Year
grouped_df = df.groupby('year_x')
# Iterate over the groups and print the subsets
for group_name, group_df in grouped_df:
    print(f"Subset for year_x {group_name}:")
    print(group_df)
    print()
```

```python
#illustration of how splitting the dataset by year impacts the IMDB Score
mean_IMDB = grouped_df['imdb_score_x'].mean()
print(mean_IMDB)
```

this shows the variation in the series or movie performance and the corresponding year

```python
# Visualisation for the result
year = [2001,2004,2005,2010,2011,2012,2014,2016,2019]
mean_IMDB = [8.700000,8.400000,8.700000,8.300000,8.500000,8.600000,8.566667,7.900000,8.850000]

plt.plot(year,mean_IMDB)
plt.xlabel('Year')
plt.ylabel('IMDB Mean Score')
plt.title('IMDB Mean Score Over Years')
plt.show()
```

The line plot shows the trend of the IMDB mean score over time. The IMDB mean score seems to increase over time, reaching a peak around 2014 and then gradually decreasing. This could be interpreted as an increase in the quality of movies over time, followed by a decline. 

```python
#Create a new feature to represent the age of the movie/series
df['title_age'] = 2023-df['year_x']
df.head()
```

```python
#Plot the relationship between the title age and IMDB score.
plt.figure(figsize=(10, 6))
plt.scatter(df['title_age'], df['imdb_score_x'])
plt.xlabel('Title Age')
plt.ylabel('IMDB Score')
plt.title('Scatter Plot for Title Age and IMDB Score')
plt.show()
```

the highest IMDB score is at the about 2 years meaning the movie or TV show is relatively new while the lowest IMDB score is between 10 years and 12.5.

```python
sns.jointplot(x="title_age", y="imdb_score_x", data=df, kind='hex', palette= 'viridis')

g = sns.jointplot(x="title_age", y="imdb_score_x", data=df, kind='hex')
g.set_axis_labels('Title Age', 'IMDb Score')

plt.show()
```


The hexbin plot provides a visual representation of the distribution of imdb_score_x for different title_age values. The color of each hexagon represents the density of the data points in that bin. Darker colors represent higher densities.
